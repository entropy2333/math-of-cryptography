## 例题

```python
# 求解逆元
n = 29
R = Integers(n)
a = R(12)
a^(-1)
```

```python
crt([1, 2, 1], [3, 5, 7])
```

```python
# 打印逆元表
n_list = list(primes(2, 50))
print(n_list)
for p in n_list:
    print(f"{'-'*50}\np: {p}")
    res = []
    for i in range(2, p):
        R = Integers(p)
        a = R(i)
        res.append((i, a^(-1)))
    print(res)
```
